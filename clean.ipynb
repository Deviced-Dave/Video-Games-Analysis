```python
for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = df[column].astype('category')
```

```python
year_null = df[df['year'].isnull()].index
```

```python
year_name_miss = df[df['year'].isnull()]['name'].to_list()
```

```python
print(year_name_miss)
```

```python
import os
import openai

from dotenv import load_dotenv
load_dotenv()

openai.api_key = os.environ["OpenAI_key"]
```

```python
model = "gpt-3.5-turbo"
answers = []

questions = [
    "Grand Theft Auto San Andreas",
    "Grand Theft Auto V"
]

conversation =[
    {"role": "system",
     "content":"You have to tell the year in which the games where published. You speak in a concise manner."},
    {"role":"user",
     "content":"Grand Theft Auto Vice city"},
    {"role":"assistant",
     "content":"2002"},
    ]

for question in year_name_miss:

    input_dict = {"role": "user",
                  "content": question}
    
    conversation.append(input_dict)  

    response = openai.chat.completions.create(
            model=model,
            messages=conversation,
            temperature=0.0,
            max_tokens=100
    )

    resp = response.choices[0].message.content
    print(resp)

    resp_dict = {"role": "assistant",
                 "content": resp}
    
    conversation.append(resp_dict)
    answers.append(resp)
```

```python
print(answers)
print(year_name_miss)
```

```python
df2 = pd.DataFrame({'name': year_name_miss, 'year1': answers})
```

```python
df2
```

```python
merged_df = pd.merge(df, df2, on='name', how='left')

merged_df['year'] = merged_df['year'].fillna(merged_df['year1'])

merged_df.drop(columns=['year1'], inplace=True)
```

```python
merged_df.info()
```

```python
merged_df = merged_df.drop_duplicates()
```

```python
merged_df.to_csv('preprocessed_data.csv', index=True)
```
